```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
np.random.seed(100)
```

```python
plt.figure(figsize=(10,5))

x = np.linspace(-6.28, 6.28, num=60)
y = np.sin(x) + np.random.normal(0, 0.1, size = x.shape[0])
plt.plot(x, np.sin(x), 'y:')
plt.plot(x, y, 'go', markersize=3)
```
